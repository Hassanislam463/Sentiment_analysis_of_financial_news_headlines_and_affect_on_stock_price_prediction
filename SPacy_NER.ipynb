```python
from spacy.pipeline import EntityRuler
from spacy.language import Language
```

```python
# Function to set custom entities
@Language.component("set_custom_ents")
def set_custom_ents(doc):
    ruler = EntityRuler(nlp)
    patterns = [{"label": "ORG", "pattern": [{"lower": symbol.lower()} for symbol in company_mapping.keys()]}]
    ruler.add_patterns(patterns)
    doc = nlp.get_pipe("ner")(doc)
    doc.ents = [ent for ent in doc.ents if ent.label_ == "ORG"]
    return doc
```

```python
# Add the custom NER rule to the pipeline after the default NER component
nlp.add_pipe("set_custom_ents", after="ner")
```

```python
# Path to the directory with dataset chunks
data_dir = r'C:\Users\hassa\OneDrive\Desktop\Datasets_to_explore\reuters'
output_dir = r'D:\Processed_data'

# Ensure output directory exists
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python
# Check which files have already been processed
processed_files = set(os.listdir(output_dir))
```

```python
# Iterate over each TSV file in the Reuters directory
for file_name in sorted(os.listdir(data_dir)):
    if file_name.endswith('.tsv') and f'processed_{file_name}' not in processed_files:
        file_path = os.path.join(data_dir, file_name)
        df = pd.read_csv(file_path, sep='\t')

        # Drop 'href' column
        if 'href' in df.columns:
            df = df.drop('href', axis=1)

        # Apply NER to the chunk in batches
        processed_docs = list(nlp.pipe(df['title'], batch_size=300))

        # Extract entities from the processed documents
        df["entities"] = [[(ent.text, ent.label_) for ent in doc.ents] for doc in processed_docs]

        # Map symbols to company names in the 'entities' column
        df["entities"] = df["entities"].apply(lambda entities: [(next((ticker for ticker, name in company_mapping.items() if name in entity), entity), label) for entity, label in entities])

        # Filter rows that don't have ORG entities
        df_org_only = df[df["entities"].apply(lambda entities: any(entity[1] == "ORG" and entity[0] in company_mapping for entity in entities))]

        # Save the processed chunk to a new file
        output_file_path = os.path.join(output_dir, f'processed_{file_name}')
        df_org_only.to_csv(output_file_path, index=False)
        print(f"Processed and saved {file_name}")
```
